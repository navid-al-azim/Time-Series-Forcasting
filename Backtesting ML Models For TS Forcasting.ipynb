**Backtesting Machine Learning Models for Time Series Forcasting**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv'
data = pd.read_csv(url, header = 0, index_col = 0, parse_dates = True)
data.head()
```

```python
data.plot()
plt.show()
```

```python
x = data.values
size = int(len(x) * .65)
train, test = x[0:size], x[size:]
print('Length of total Dataset: ' + str(len(x)))
print('-'*50)
print('Length of Train Dataset: ' + str(len(train)))
print('-'*50)
print('Length of Test Dataset: ' + str(len(test)))
```

```python
plt.plot(train)
plt.plot([None for i in train] + [x for x in test])
plt.show()
```

```python
n_train = 500
n_record = len(x)

for i in range(n_train, n_record):
    trains, tests = x[0:i], x[i:i+1]
    print('Length of Train: ' + str(len(trains)) + ' && Length of Test: '+str(len(tests)))
```
